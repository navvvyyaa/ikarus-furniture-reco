# 01 — Data Analytics\nExplore dataset and export analytics JSON.

```python
import pandas as pd, json
from collections import Counter
df = pd.read_csv('../data/raw/intern_data_ikarus.csv')
brands = Counter(df.get('brand').fillna('Unknown'))
cats = Counter([str(c).split(',')[0].strip() for c in df.get('categories').fillna('Unknown')])
prices = pd.to_numeric(df.get('price').astype(str).str.replace(r'[^\d.]','',regex=True), errors='coerce').dropna().tolist()
summary = {'total_products': int(df.shape[0]), 'brand_counts': brands.most_common(20), 'category_counts': cats.most_common(20), 'price_summary': {'min': min(prices) if prices else None, 'max': max(prices) if prices else None, 'avg': (sum(prices)/len(prices)) if prices else None}}
open('../backend/app/data_analytics.json','w').write(json.dumps(summary, indent=2))
summary
```
